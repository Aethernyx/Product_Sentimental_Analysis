```python
import numpy as np
import pandas as pd
import re
from bs4 import BeautifulSoup
import selenium
from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
from selenium.common.exceptions import TimeoutException
from selenium.webdriver.common.keys import Keys
import time

from selenium.common.exceptions import NoSuchElementException
```

```python
chromedriver = '/Users/alex/downloads/chromedriver 3'
browser = webdriver.Chrome(chromedriver)
```

```python
def scroll_down():
    body = browser.find_element_by_css_selector('body')
    body.send_keys(Keys.PAGE_DOWN)

def get_reviews(url, reviews_to_scrape):
    browser.get(url)
    try:
        browser.find_element_by_xpath('//*[@id="acrCustomerReviewLink"]').click()
    except NoSuchElementException:
        print('This prodcut has no reviews!')
    
    for i in range(10):
        scroll_down()
    
    review_text = []
    review_score = []
    review_date = []
    review_location = []
    
    
    try:
        browser.find_element_by_xpath('//*[@id="cr-pagination-footer-0"]/a').click()
    except NoSuchElementException:
        print('Not enough reviews available for this product!')
        pass
        
    review_count = 0
        
    while review_count < reviews_to_scrape:
        
        for i in range(10):
            scroll_down()
        
        page_source_data = browser.page_source
        soup = BeautifulSoup(page_source_data, 'html.parser')
        list_of_review_text = soup.find_all(attrs={'class': 'a-size-base review-text review-text-content'})
        list_of_review_ratings = soup.find_all(attrs={'class': 'a-icon-alt'})[3:]
        list_of_dates = soup.find_all(attrs={'class': 'a-size-base a-color-secondary review-date'})[2:]
            
        for i in range(len(list_of_reviews)):
            review_text.append(list_of_review_text[i].text)
            review_score.append(list_of_review_ratings[i].text[:3])
            review_date.append(list_of_dates[i].text.split('on ')[-1])
            review_location.append(list_of_dates[i].text.split('in ')[1].split('on ')[0])
        try:    
            browser.find_element_by_class_name('a-last').click()
        except NoSuchElementException:     
            print('Scraped all available reviews for this product!')
            break
        
            
        review_count += len(list_of_review_text)
    df = pd.DataFrame()
    df['review_date'] = review_date
    df['review_location'] = review_location
    df['review_text'] = review_text
    df['review_score'] = review_score 
    
    return df
```

```python
reviews = get_reviews('https://www.amazon.com/48-Laws-Power-Robert-Greene/dp/0140280197/ref=sr_1_2?dchild=1&keywords=48+Laws+Power&qid=1600206674&sr=8-2', 50)
```
